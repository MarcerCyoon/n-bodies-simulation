```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation


# Holds the Particle class, which is instantiated for each particle in the simulation.
class Particle:
    def __init__(self, vel, pos, mass):
        # vel, pos, acc are all 2-dimensional vectors.
        self.vel = vel
        self.pos = pos
        self.acc = np.array([0.0, 0.0])
        self.mass = mass
       
            
    # The applyForce() function calculates the acceleration the particle will receive based on the
    # gravitational force. This is then stored in the particle's acceleration.
    def applyForce(self, particle):
        x_diff = particle.pos[0] - self.pos[0]
        y_diff = particle.pos[1] - self.pos[1]
        dist = (x_diff ** 2 + y_diff ** 2) ** (1 / 2)
        
        # NOTE: These are the vectorized forms of Newton's Second Law, hence the r^3 at the bottom and the extra
        # diff multiplier.
        x_force = (G_CONST * particle.mass / (dist ** 3)) * x_diff
        y_force = (G_CONST * particle.mass / (dist ** 3)) * y_diff
        
        self.acc[0] = x_force
        self.acc[1] = y_force
        
        
    # The update() function moves the particle one time-step up, effectively "updating" it.    
    def update(self):
        print("Vel: {}, Acc: {}".format(self.vel, self.acc))
        self.vel[0] = self.acc[0] + self.vel[0]
        self.vel[1] = self.acc[1] + self.vel[1]
        
        self.pos[0] = self.vel[0] + self.pos[0]
        self.pos[1] = self.vel[1] + self.pos[1]

# Setting some constants here.
# A two-body simulation is what we are currently demonstrating.
# Time Constant tells us how many steps to run.
# A 1000 mass is assigned to both objects, for now, to simplify the prototype.
NUM_PARTICLES = 2
TIME = 100
MASS_ONE = 1000
MASS_TWO = 1000

# Using a slightly bigger G so that the forces apply more quickly, for the sake of nicer animations.
G_CONST = 6.67 * (10 ** -5)
```

```python
# The PARTICLES list stores all the particles in the simulation.
PARTICLES = [Particle([0, 0], [0, 0], MASS_ONE), Particle([0, 0], [0, 0], MASS_TWO)]

# Setting the Initial Conditions of the particles.
vel = np.array([0.0, 0.0])
pos = np.array([6.0, -1.0])
vel_two = np.array([0.0, 0.0])
pos_two = np.array([-4.0, 3.0])
```

```python
# Setting plot x and y axes
plt.xlim(-40, 40)
plt.ylim(-40, 40)

# Initialize Particles
PARTICLES[0] = Particle(vel, pos, MASS_ONE)
PARTICLES[1] = Particle(vel_two, pos_two, MASS_TWO)

plt.plot(PARTICLES[0].pos[0], PARTICLES[0].pos[1], 'bo')
plt.plot(PARTICLES[1].pos[0], PARTICLES[1].pos[1], 'bo')
plt.savefig('gif/image0.png', format='png') # Save current plot as an image to be used to make gif
plt.show()

for i in range(TIME):
    PARTICLES[0].applyForce(PARTICLES[1])
    PARTICLES[1].applyForce(PARTICLES[0])
    PARTICLES[0].update()
    PARTICLES[1].update()
    print("Particle 1: ({}, {}); Particle 2: ({}, {})".format(PARTICLES[0].pos[0], PARTICLES[0].pos[1], PARTICLES[1].pos[0], PARTICLES[1].pos[1]))
    plt.xlim(-40, 40)
    plt.ylim(-40, 40)
    plt.plot(PARTICLES[0].pos[0], PARTICLES[0].pos[1], 'bo')
    plt.plot(PARTICLES[1].pos[0], PARTICLES[1].pos[1], 'bo')
    fileName = 'gif/image' + str(i + 1) + '.png'
    plt.savefig(fileName, format='png')
    plt.show()
```
